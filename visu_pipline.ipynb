```python
def show_image(image, title, cmap_type="gray"):
    image = cv2.colorChange(image, cv2.COLOR_BGR2RGB)
    plt.title(title)
    plt.imshow(image, cmap=cmap_type)
```

```python
image1 = cv2.imread('./test_data/test_images/straight_lines1.jpg')
thresh_img = thresholding.combine_thresh(image1)
show_image(thresh_img, "threshold")
```

```python
def displayVideo(path):
    video = cv2.VideoCapture(path)
    while(True):
        retVal, frame = video.read()
        if(not retVal):
            video.set(cv2.CAP_PROP_POS_FRAMES, 0)
            continue
        
        
        cv2.imshow('raw', frame)
        cv2.imshow('out', thresholding.combine_thresh(frame))

        if(cv2.waitKey(30) == ord('q')):
            break
    video.release()
    cv2.destroyAllWindows()
```
